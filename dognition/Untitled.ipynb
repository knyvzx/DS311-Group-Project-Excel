```python
import pandas as pd
df1 = pd.read_excel("dognition_data_aggregated_by_dogid.xlsx")
df1.head()
```

```python
df2 = pd.read_excel('dognition_data_no_aggregation_with_zip_code_correction.xlsx')

df2
```

```python
columns_df1 = df1.columns
print("Column names for df1:", columns_df1)
```

```python
columns_df2 = df2.columns
print("Column names for df2:", columns_df2)
```

```python
# Select and display only the 'Dog ID' column from df1
dog_id_column_df1 = df1['Dog ID']
print(dog_id_column_df1.head())
# Select and display only the 'Dog ID' column from df2
dog_id_column_df2 = df2['Dog_ID']
print(dog_id_column_df2.head())

```

```python
# Merge the dataframes on multiple columns
merged_df = pd.merge(df1, df2, left_on=['User ID','Dog ID', 'Gender', 'Birthday', 'Breed', 'Breed_Type', 'Breed_Group', 'Weight', 'Dog_Fixed', 'DNA_Tested', 'Dimension', 'Sign_in_Count', 'Max_Dogs', 'Membership_ID', 'Subscribed', 'City', 'State', 'Zip', 'Country', 'Exclude', 'Free_Start_User', 'Last_Active_At', 'Membership_Type'], right_on=['User_ID','Dog_ID', 'Gender', 'Birthday', 'Breed', 'Breed_Type', 'Breed_Group', 'Weight', 'Dog_Fixed', 'DNA_Tested', 'Dimension', 'Sign_in_Count', 'Max_Dogs', 'Membership_ID', 'Subscribed', 'City', 'State', 'Zip', 'Country', 'Exclude', 'Free_Start_User', 'Last_Active_At', 'Membership_Type'], how='inner', suffixes=('_df1', '_df2'))

# Display the merged dataframe
merged_df.head()
```

```python
columns_merged_df = merged_df.columns
print("Column names for merged_df:", columns_merged_df)
```

```python
# Sort the merged DataFrame by the 'Dog_Fixed' column
merged_df_sorted = merged_df.sort_values(by='Dog_Fixed')

# Display the sorted DataFrame
merged_df_sorted.head()

```

```python
# Filter the merged DataFrame to show only dogs that are fixed
fixed_dogs_df = merged_df[merged_df['Dog_Fixed'] == True]

# Display the filtered DataFrame
fixed_dogs_df.head()
```

```python
# Filter the merged DataFrame to show only dogs that are not fixed
not_fixed_dogs_df = merged_df[merged_df['Dog_Fixed'] == False]

# Display the filtered DataFrame
not_fixed_dogs_df.head()
```

```python
# Create a DataFrame for male dogs
male_dogs_df = merged_df[merged_df['Gender'] == 'male']

# Create a DataFrame for female dogs
female_dogs_df = merged_df[merged_df['Gender'] == 'female']

# Display the DataFrames
print("Male Dogs:")
print(male_dogs_df.head())

print("\nFemale Dogs:")
print(female_dogs_df.head())

```

```python
# Sort male and female dogs in df1 by 'Breed'
sorted_male_dogs_df1 = df1[df1['Gender'] == 'male'].sort_values(by='Breed')
sorted_female_dogs_df1 = df1[df1['Gender'] == 'female'].sort_values(by='Breed')

# Sort male and female dogs in df2 by 'Breed'
sorted_male_dogs_df2 = df2[df2['Gender'] == 'male'].sort_values(by='Breed')
sorted_female_dogs_df2 = df2[df2['Gender'] == 'female'].sort_values(by='Breed')

# Display the sorted DataFrames
print("Male Dogs in df1:")
print(sorted_male_dogs_df1.head())

print("\nFemale Dogs in df1:")
print(sorted_female_dogs_df1.head())

print("\nMale Dogs in df2:")
print(sorted_male_dogs_df2.head())

print("\nFemale Dogs in df2:")
print(sorted_female_dogs_df2.head())
```

```python
# Display the summary counts after sorting
print("Summary for Male Dogs in df1:")
print(sorted_male_dogs_df1['Breed'].value_counts())

print("\nSummary for Female Dogs in df1:")
print(sorted_female_dogs_df1['Breed'].value_counts())

print("\nSummary for Male Dogs in df2:")
print(sorted_male_dogs_df2['Breed'].value_counts())

print("\nSummary for Female Dogs in df2:")
print(sorted_female_dogs_df2['Breed'].value_counts())
```

```python
# Get the summary counts for male and female dogs in df1
male_counts_df1 = sorted_male_dogs_df1['Breed'].value_counts()
female_counts_df1 = sorted_female_dogs_df1['Breed'].value_counts()

# Get the summary counts for male and female dogs in df2
male_counts_df2 = sorted_male_dogs_df2['Breed'].value_counts()
female_counts_df2 = sorted_female_dogs_df2['Breed'].value_counts()

# Combine the counts for male and female dogs
combined_counts_df1 = pd.DataFrame({'Male': male_counts_df1, 'Female': female_counts_df1})
combined_counts_df2 = pd.DataFrame({'Male': male_counts_df2, 'Female': female_counts_df2})

# Display the combined counts
print("Summary Counts for df1:")
print(combined_counts_df1.head())

print("\nSummary Counts for df2:")
print(combined_counts_df2.head())
```

```python
# Calculate the ratio of males to females for each dog breed in df1
combined_counts_df1['Ratio (Male/Female)'] = combined_counts_df1['Male'] / combined_counts_df1['Female']

# Calculate the ratio of males to females for each dog breed in df2
combined_counts_df2['Ratio (Male/Female)'] = combined_counts_df2['Male'] / combined_counts_df2['Female']

# Display the calculated ratios
print("Ratios for df1:")
print(combined_counts_df1)

print("\nRatios for df2:")
print(combined_counts_df2)
```

```python
# Display the calculated ratios for df1 without NaN values
print("Ratios for df1:")
print(combined_counts_df1.dropna())

# Display the calculated ratios for df2 without NaN values
print("\nRatios for df2:")
print(combined_counts_df2.dropna())
```

```python
# Sort the DataFrame by the calculated ratios in df1
sorted_ratios_df1 = combined_counts_df1.sort_values(by='Ratio (Male/Female)', ascending=False)

# Sort the DataFrame by the calculated ratios in df2
sorted_ratios_df2 = combined_counts_df2.sort_values(by='Ratio (Male/Female)', ascending=False)

# Display the sorted DataFrames
print("Sorted Ratios for df1:")
print(sorted_ratios_df1.dropna())

print("\nSorted Ratios for df2:")
print(sorted_ratios_df2.dropna())
```
