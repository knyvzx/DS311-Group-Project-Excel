```python
import pandas as pd

# Load the datasets
df = pd.read_excel('dognition/dognition_data_aggregated_by_dogid.xlsx')
df1 = pd.read_excel('dognition/dognition_data_no_aggregation_with_zip_code_correction.xlsx')

# 1. Display basic information about the DataFrames
print("\nBasic Info about df:")
print(df.info())

print("\nBasic Info about df1:")
print(df1.info())

# 2. Show summary statistics for numeric columns
print("\nSummary Statistics for df:")
print(df.describe())

print("\nSummary Statistics for df1:")
print(df1.describe())

# 3. Check for missing values
print("\nMissing Values in df:")
print(df.isnull().sum())

print("\nMissing Values in df1:")
print(df1.isnull().sum())

# 4. Explore unique values in categorical columns
print("\nUnique Values in 'Breed' column of df:")
print(df['Breed'].unique())

print("\nUnique Values in 'Gender' column of df1:")
print(df1['Gender'].unique())

# 5. Analyze correlation between columns
correlation_matrix = df.corr()
print("\nCorrelation Matrix for df:")
print(correlation_matrix)

# Note: Adjust the column names as needed based on your specific analysis requirements.
```
